```python
%load_ext rpy2.ipython
%matplotlib inline
import pandas as pd
import numpy as np
from pylab import pcolor, gca
```

```python
%%R -o all_pca,all_krz_proj,all_krz,all_mantel

library(Morphometrics)

load('/home/walrus/Dropbox/labbio/cov_bayes_data/Rdatas/nwm.Rdata')

all_pca <- PCAsimilarity(nwm.matrices, repeat.vector=rep(1,16), num.cores=10)
all_krz_proj <- KrzProjection(nwm.matrices, num.cores=10)
all_krz <- KrzCor(nwm.matrices,repeat.vector=rep(1,16), num.cores=10)
all_mantel <- MantelCor(nwm.matrices,repeat.vector=rep(1,16), num.cores=10)$correlations
```

```python
pca = np.array(all_pca)
krz = np.array(all_krz)
krz_proj = np.array(all_krz_proj)
mantel = np.array(all_mantel)
```
