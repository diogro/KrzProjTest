```python
%load_ext rpy2.ipython
%matplotlib inline
import pandas as pd
import numpy as np
from pylab import pcolor, gca
```

```python
cors = []
num_matrices = 11
matrix_size = 27

for i in range(11):
    cors += [pd.read_csv('../cor.matrix.csv', sep=',', header=None, skiprows=1 + i*(matrix_size + 2), nrows=matrix_size, comment=' ')]
```

```python
%Rpush cors
```

```python
%%R -o all_proj,all_krz,all_pcasim

library(Morphometrics)

all_pcasim <- laply(1:11, function (i) laply(1:11, function (j) PCAsimilarity(cors[i,,], cors[j,,])))
all_krz <- laply(1:11, function (i) laply(1:11, function (j) KrzCor(cors[i,,], cors[j,,])))
all_proj <- laply(1:11, function (i) laply(1:11, function (j) KrzProjection(cors[i,,], cors[j,,])$total.variation))
```

```python
pca = np.array(all_pcasim)
krz = np.array(all_krz)
krz_proj = np.array(all_proj)
```

## Krzanowski Projection

```python
pd.DataFrame(krz_proj)
```

```python
pd.DataFrame(krz_proj - krz_proj.T)
```

```python
pcolor(krz_proj - krz_proj.T)
gca().invert_yaxis()
```
